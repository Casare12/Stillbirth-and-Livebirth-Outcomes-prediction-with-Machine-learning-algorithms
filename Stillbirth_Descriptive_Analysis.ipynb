```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('df1.csv')
data.head()
```

```python
data.isna().sum()
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Define the mapping of x-labels to new labels
label_mapping = {0: 'SVD', 1: 'C/S', 2: 'SVD & C/S'}

# Map the 'Deliverymethod' column to the new labels
data['Deliverymethod'] = data['Deliverymethod'].map(label_mapping)

# Stacked Bar Chart with modified legend and x-labels
sns.countplot(x='Deliverymethod', hue='Status', data=data)
plt.title("The method of Delivery and the Status results")
plt.legend(title="Status", labels=["0=Still Birth", "1=Live Birth"])
plt.show()
```

```python
# Box Plots
continuous_cols = ['age', 'gestation', 'Hemamount', 'Weight']

# Define the mapping of x-labels to new labels
label_mapping = {0: 'Livebirth', 1: 'Stillbirth'}

# Map the values in the 'Status' column to their corresponding labels
data['Status'] = data['Status'].map(label_mapping)

for col in continuous_cols:
    sns.boxplot(x='Status', y=col, data=data)
    plt.title(f"{col} vs. Status")
    plt.show()
```

```python
# Histograms
for col in continuous_cols:
    sns.histplot(x=col, data=data, hue='Status', kde=True)

    # Define a dictionary to map values with 0 as Stillbirth and 1 as Livebirth
    status_mapping = {0: 'Stillbirth', 1: 'Livebirth'}

    # Create a new column with the mapped labels
    data['Status_Label'] = data['Status'].map(status_mapping)

    plt.title(f"{col} Distribution by Status")
    plt.show()
```

```python
# Subset the data to include only the columns of interest
columns_of_interest = ['age', 'Hemamount', 'Fetalheart', 'Deliverymethod', 'Hxob']
subset_data = data[columns_of_interest]

# Calculate the correlation matrix for the subset data
corr_matrix_subset = subset_data.corr()

# Create a correlation heatmap
sns.heatmap(corr_matrix_subset, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap for Age, Hemamount, fetalheart, Deliverymethod, Hxob")
plt.show()
```
